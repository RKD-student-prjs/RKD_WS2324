```python

import json
import re

new_data = []

### functions

def create_fach(fach, prof):
        final_dict = {
            "fach": fach,
            "professoren": prof 
        }
        #print(json.dumps(final_dict, indent=4))
        return final_dict
    
    
def check_fach(fach):
    
    if ";" in fach:
        #print(fach)
        return True
    
def check_link(link):
    pattern = 'http'
    domain = 'https://www.th-koeln.de'

    if pattern not in link:
        full_link = domain + link
        return full_link

    return link
    
        
def fix_link_list(links_str):
    
    new_link_list = []
    
    split_links = links_str.split('\n')
    for link in split_links:
        clean_link = re.sub("'|\[|\]|\s+/g", "", link)
        checked_link = check_link(clean_link)
        final_link = checked_link.replace(" ", "")
        new_link_list.append(final_link)
        
    return new_link_list
        
    
    
### Opening JSON file
with open('./src/data/data.json') as json_file:
    raw_data = json.load(json_file)
    
    
    for data in raw_data:
        #print(json.dumps(data, indent=4))
        
        
        for key, value in data.items():
            
            new_themen_dict = {
                "themengebiet": data["themengebiet"],
                "faecher": []
            }
            if key == "faecher":
                for fach_dict in value:
                    #print(fach_dict)

                    # fix links in profs
                    for profs in fach_dict["professoren"]:
                        profs["links"] = fix_link_list(profs["links"])     
                        
                    # fix multiple "fach1;fach2"
                    #if check_fach(fach_dict["fach"]):
                        #for fach in fach_dict["fach"].split(";"):
                            #create_fach(fach, fach_dict["professoren"])
                    #else:
                    new_themen_dict["faecher"].append(fach_dict)
            
        new_data.append(new_themen_dict)

                        
                        
    print(json.dumps(new_data, indent=4))
    
with open("./src/data/sample.json", "w") as outfile:
    json.dump(new_data, outfile)
```
